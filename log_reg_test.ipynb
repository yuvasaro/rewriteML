```python
import numpy as np
from log_reg import f_wb, zscore_normalize, compute_cost, compute_gradient, gradient_descent, plot_J_hist

np.set_printoptions(precision=3)
```

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])
```

```python
# gradient descent settings
n = X_train.shape[1]
w_in = np.zeros((n,))
b_in = 0
alpha = 1e-1

# run gradient descent
w_final, b_final, J_hist = gradient_descent(
    X_train, 
    y_train, 
    w_in, 
    b_in, 
    alpha, 
    model=f_wb, 
    cost_func=compute_cost, 
    grad_func=compute_gradient, 
    num_iter=10000
)

print(f"Results: w = {w_final}, b = {b_final:0.3e}")
```

```python
plot_J_hist(J_hist)
```
